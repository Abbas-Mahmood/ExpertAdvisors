```python
#Open Buy, if 10 pip down close buy and open sell. if sell 10 pip up close sell open buy,
#Take profit also at 10 pips - then continue in same direction
Profit = [0]
PipProfit = 0
EntryPrice = XAUUSD['Ask'].iloc[0]
TradeType = ['Buy']
for i in range(len(XAUUSD)):
    if TradeType[-1] == 'Buy':
        if XAUUSD['Ask'].iloc[i] < EntryPrice-1:
            Profit.append(XAUUSD['Bid'].iloc[i]-EntryPrice)
            PipProfit += XAUUSD['Bid'].iloc[i]-EntryPrice
            EntryPrice = XAUUSD['Bid'].iloc[i]
            TradeType.append('Sell')
        elif XAUUSD['Ask'].iloc[i] > EntryPrice+1:
            Profit.append(XAUUSD['Bid'].iloc[i]-EntryPrice)
            PipProfit += XAUUSD['Bid'].iloc[i]-EntryPrice
            EntryPrice = XAUUSD['Ask'].iloc[i]
            TradeType.append('Buy')
    elif TradeType[-1] == 'Sell':
        if XAUUSD['Bid'].iloc[i] > EntryPrice+1:
            Profit.append(EntryPrice - XAUUSD['Ask'].iloc[i])
            PipProfit += EntryPrice - XAUUSD['Ask'].iloc[i]
            EntryPrice = XAUUSD['Ask'].iloc[i]
            TradeType.append('Buy')
        elif XAUUSD['Bid'].iloc[i] < EntryPrice-1:
            Profit.append(EntryPrice - XAUUSD['Ask'].iloc[i])
            PipProfit += EntryPrice - XAUUSD['Ask'].iloc[i]
            EntryPrice = XAUUSD['Bid'].iloc[i]
            TradeType.append('Sell')
```
